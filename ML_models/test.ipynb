# DATA


```python
new_test=pd.read_table("/home/clau/Escritorio/microbia_interaction-main/data/TEST_all.txt")
X_test=new_test.iloc [:, 1:6282].values
inter=new_test.iloc[:, 0].values
```

# Load the training XGBoost models

```python
from joblib import load
foldermodels="/home/clau/Escritorio/microbia_interaction-main/ML_models/models/XGBoost/"

F1 = load(foldermodels+'XGBoost_F1.joblib')
F2 = load(foldermodels+'XGBoost_F2.joblib')
F3 = load(foldermodels+'XGBoost_F3.joblib')
F4 = load(foldermodels+'XGBoost_F4.joblib')
```

# test experimental interactions

```python
#F1
y_pred1 = F1.predict(X_test)

#F2
y_pred2 = F2.predict(X_test)

#F3
y_pred3 = F3.predict(X_test)

#F4
y_pred4 = F4.predict(X_test)
```

Save in a new file the results

```python
file_new = open("interaction_new_paper.txt" , "w")
for i in range(0,13572):
    if(y_pred2[i]+y_pred1[i]+y_pred4[i]+y_pred3[i])<= 2:
        file_new.write(inter[i]+" CO "+str(y_pred2[i]+y_pred1[i]+y_pred4[i]+y_pred3[i])+ "\n") 
        print(inter[i]+" CO")
    
    if(y_pred2[i]+y_pred1[i]+y_pred4[i]+y_pred3[i])> 2:
        file_new.write(inter[i]+" CF "+str(y_pred2[i]+y_pred1[i]+y_pred4[i]+y_pred3[i])+"\n")
        print(inter[i]+" CF")
    
file_new.close()
```
